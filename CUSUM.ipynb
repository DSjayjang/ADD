```python
start_point = []

for i in range(4):
    phase1_data = gas_random.values[range(0, 100), i]
    phase2_data = gas_random.values[range(100, gas_random.shape[0]), i]

    phase2_cusum = spc(phase1_data,newdata = phase2_data) + cusum() +  rules()


    summary = phase2_cusum.summary[0]
    cplus = summary["values"][0]       # C⁺ (상한 누적합)
    cminus = summary["values"][1]      # C⁻ (하한 누적합)
    ucl = summary["ucl"]               # Upper control limit
    lcl = summary["lcl"]               # Lower control limit

    upper_violations = [i for i, val in enumerate(cplus) if val > ucl]
    lower_violations = [i for i, val in enumerate(cminus) if val < lcl]

    if upper_violations or lower_violations:
        start_point = min(upper_violations + lower_violations)
        print("Start point (in Phase 2):", start_point)
    else:
        print("No violations detected.")
```
