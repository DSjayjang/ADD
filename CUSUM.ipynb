```python
# 우선 univariate에 대해서만.
h = 0
phase1_len = 100

phase1_data = gas_random.values[range(0, phase1_len), 0]
phase2_data = gas_random.values[range(phase1_len, gas_random.shape[0]), 0]
```

```python
class uniCUSUM:
    def __init__(self, phase1_len = 100, delta = 0.5, h = 5):
        self.delta = float(delta)
        self.h = float(h)
        self.phase1_len = phase1_len

        self.s_prev_pos_ = 0.0
        self.s_prev_neg_ = 0.0

        self.mu0_ = None
        self.sigma_ = None
        
        # self.s_pos_history = []
        # self.s_neg_history = []

    def fit(self, X: np.array):
        # phase 1
        self.phase1_data = X[:self.phase1_len]
        self.mu0_ = np.mean(self.phase1_data)
        self.sigma_ = np.std(self.phase1_data, ddof = 1)

    def update(self, x_t):
        # phase 2
        s_t_pos = max(0.0, self.s_prev_pos_ + (x_t - self.mu0_ - self.delta/2) / self.sigma_)
        s_t_neg = max(0.0, self.s_prev_neg_ + (self.mu0_ - x_t - self.delta/2) / self.sigma_)

        # 벡터 누적합 업데이트
        self.s_prev_pos_ = s_t_pos
        self.s_prev_neg_ = s_t_neg

        return s_t_pos, s_t_neg
    
    def reset(self):
        # 벡터 누적합 리셋
        self.s_prev_pos_ = 0.0
        self.s_prev_neg_ = 0.0

    # def plot(self, time_labels=None):
    #     """누적 통계량 시각화"""
    #     plt.figure(figsize=(12, 5))

    #     x = np.arange(len(self.s_pos_history))
    #     if time_labels is not None and len(time_labels) == len(x):
    #         x = time_labels

    #     plt.plot(x, self.s_pos_history, label='CUSUM Positive', color='blue')
    #     plt.plot(x, self.s_neg_history, label='CUSUM Negative', color='red')
    #     plt.axhline(self.h, color='black', linestyle='--', label='Control Limit (+h)')
    #     plt.axhline(-self.h, color='gray', linestyle='--', label='Control Limit (-h)')

    #     plt.title("CUSUM Control Chart")
    #     plt.xlabel("Time" if time_labels is not None else "Index")
    #     plt.ylabel("CUSUM Statistic")
    #     plt.legend()
    #     plt.grid(True)
    #     plt.tight_layout()
    #     plt.show()
```

```python
start = []
violoation_lower, violoation_upper = [], []

cusum = uniCUSUM(phase1_len)
cusum.fit(phase1_data)

for idx, x in enumerate(phase2_data):
    s_p, s_n = cusum.update(x)

    if s_p > h:
        start_idx = phase1_len + idx
        violoation_upper.append(start_idx)
        cusum.reset()

    if s_n > h:
        start_idx = phase1_len + idx
        violoation_lower.append(start_idx)
        cusum.reset()

start_point = min(violoation_lower + violoation_upper)
print(start_point)
print(time_labels[start_point])
```
